```python
def sum14(* para):
    print(para)
    sum1 = 0
    if para[0] == 'add':
        for num in range(1,len(para)):
            sum1 += para[num]
    elif para[0] == 'sub':
         for num in para[1:]:
            sum1 -= num
    elif para[0] == 'mul':
         for num in para[1:]:
            sum1 *= num
    else:
         for num in para[1:]:
            sum1 /= num
    print(sum1)

sum14('add',2,4,5,6,7,8)
sum14('mul',2,3,4,5,6,7,8)
```

```python
def sum14(opt,* para):
    sum2 = 1
    if opt == 'add':
        sum2 = sum2 - 1
        for num in para:
            sum2 += num
    elif opt == 'sub':
        sum2 = sum2 - 1
        for num in para:
            sum2 -= num
    elif opt == 'mul':
        for num in para:
            sum2 *= num
    else:
        for num in para:
            sum2 /= num
    print(sum2)

sum14('add',2,4,5,6,7)
sum14('mul',2,3,4,5,6,7,8)
```

```python
class Calculator:    # 파이썬의 생성자는 하나만 생성 가능하다.
    
    def __init__(self,f,s):   # 멤버 필드를 초기화하기 위한 생성자격 메소드 __init__ (_2개씩)
        self.f = f
        self.s = s

    def addNum(self):
        self.res = self.f + self.s
        return self.res

    def divNum(self):
        self.res = self.f / self.s
        return self.res
```

```python
calc = Calculator(0,20)
print(calc.f)
res = calc.addNum()
print(res)
res = calc.divNum()
print(res)      # 0 을 나누는 것을 '부정' 이라고 한다.
```
